```python
import streamlit as st
import re
from pathlib import Path
from typing import List

from langchain_community.document_loaders import PDFPlumberLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_core.vectorstores import InMemoryVectorStore
from langchain_ollama import OllamaEmbeddings
from langchain_core.prompts import ChatPromptTemplate
from langchain_ollama.llms import OllamaLLM
from langchain_core.documents import Document
```

```python
deepseek = OllamaLLM(model = "deepseek-r1:8b")

result = deepseek.invoke("Who is jonas in the Netflix dark series?")

print(result)
```

```python
file_path = '../pdf/eng.pdf'
pdf = PDFPlumberLoader(file_path)

data = pdf.load()
len(data)
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=100,
    separators=["\n\n", "\n", ".", "!", "?", ",", " "],
    add_start_index=True
)

docs = text_splitter.split_documents(data)
```

```python
len(docs)
```

```python
# show raw text of docs
for doc in docs:
    print(doc.page_content)
```

```python
embeddings = OllamaEmbeddings(model = "deepseek-r1:8b")
vector_store = InMemoryVectorStore(embeddings)
```

```python
embeddings.model
```

```python
vector_store.add_documents(docs)
```

```python
query = "does Everyone charged with a penal offence has the right to be presumed innocent until proved guilty?"
retrieved_documents = vector_store.similarity_search(query, k=5)

print(retrieved_documents)
```

```python
context = "\n".join([doc.page_content for doc in retrieved_documents])
```
